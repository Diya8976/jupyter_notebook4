```python
import requests
def download(url, filename):
    response = requests.get(url)
    if response.status_code == 200:
        with open(filename, "wb") as f:
            f.write(response.content)
        print(f"{filename} download successfully.")
    else:
        print(f"failed to download {filename}. status code: {response.status_code}")
```

```python
from sklearn.metrics import accuracy_score, f1_score, jaccard_score, log_loss
import pandas as pd
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
LR = LogisticRegression(solver='liblinear')
LR.fit(X_train, Y_train)

lr_predictions = LR.predict(X_test)
lr_predict_proba = LR.predict_proba(X_test)

from sklearn.neighbors import KNeighborsClassifier
KNN = KNeighborsClassifier(n_neighbors=4)
KNN.fit(X_train, Y_train)

knn_predictions = KNN.predict(X_test)
knn_predict_proba = KNN.predict_proba(X_test)

from sklearn.tree import DecisionTreeClassifier
Tree = DecisionTreeClassifier()
Tree.fit(X_train, Y_train)

tree_predictions = Tree.predict(X_test)
tree_predict_proba = Tree.predict_proba(X_test)

SVM = SVC(probability=True)
SVM.fit(X_train, Y_train)

svm_predictions = SVM.predict(X_test)
svm_predict_proba = SVM.predict_proba(X_test)

lr_accuracy = accuracy_score(Y_test, lr_predictions)
lr_f1 = f1_score(Y_test, lr_predictions, average="weighted")
lr_jaccard = jaccard_score(Y_test, lr_predictions, average="weighted")
lr_log_loss = log_loss(Y_test, lr_predict_proba)

knn_accuracy = accuracy_score(Y_test, knn_predictions)
knn_f1 = f1_score(Y_test, knn_predictions, average="weighted")
knn_jaccard = jaccard_score(Y_test, knn_predictions, average="weighted")
knn_log_loss = log_loss(Y_test, knn_predict_proba)

tree_accuracy = accuracy_score(Y_test, tree_predictions)
tree_f1 = f1_score(Y_test, tree_predictions, average="weighted")
tree_jaccard = jaccard_score(Y_test, tree_predictions, average="weighted")
tree_log_loss = log_loss(Y_test, tree_predict_proba)

svm_accuracy = accuracy_score(Y_test, svm_predictions)
svm_f1 = f1_score(Y_test, svm_predictions, average="weighted")
svm_jaccard = jaccard_score(Y_test, svm_predictions, average="weighted")
svm_log_loss = log_loss(Y_test, svm_predict_proba)

metrics = {
    "Model": ["Logistic Regression", "KNN", "Decision Tree", "SVM"],
    "Accuracy": [lr_accuracy, knn_accuracy, tree_accuracy, svm_accuracy],
    "Jaccard Index": [lr_jaccard, knn_jaccard, tree_jaccard, svm_jaccard],
    "F1 Score": [lr_f1, knn_f1, tree_f1, svm_f1],
    "Log Loss": [lr_log_loss, knn_log_loss, tree_log_loss, svm_log_loss]
}

metrics_df = pd.DataFrame(metrics)

print(metrics_df)
```
